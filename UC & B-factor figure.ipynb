```python
def b_factor_fetcher(filename):
    data = subprocess.check_output(["phenix.b_factor_statistics", filename])
    shortname = filename.split("/")[1]
    data_string = str(data)
    data_lines = data_string.split("\\n")
    for line in data_lines:
        if "Overall" in line:
            avg_bfac = line.split("    ")[1].split("  ")[2]
            print(avg_bfac)
    return [shortname, avg_bfac]
```

```python
b_factor_fetcher(files[0])
```

```python
avg_b_factors = {}
for file in files:
    name, b_fac = b_factor_fetcher(file)
    avg_b_factors[name] = b_fac
```

# Unit Cell Expansion Figure

```python
#calculate SEM based on std dev and number of crystals for each dataset
SEM_TD = 0.10/np.sqrt(9464)

SEM_20ns_light = 0.13/np.sqrt(15253)
SEM_20us_light = 0.10/np.sqrt(13681)
SEM_200us_light = 0.11/np.sqrt(11931)

SEM_20ns_dark2 = 0.09/np.sqrt(14454)
SEM_20us_dark2 = 0.09/np.sqrt(13893)
SEM_200us_dark2 = 0.11/np.sqrt(11950)

#create 95% confidence intervals
CI_true_dark = st.norm.interval(alpha=0.95, loc=79.45, scale=SEM_TD)

CI_20ns_light = st.norm.interval(alpha=0.95, loc=79.52, scale=SEM_20ns_light)
CI_20us_light = st.norm.interval(alpha=0.95, loc=79.52, scale=SEM_20us_light)
CI_200us_light = st.norm.interval(alpha=0.95, loc=79.51, scale=SEM_200us_light)

CI_20ns_dark2 = st.norm.interval(alpha=0.95, loc=79.44, scale=SEM_20ns_dark2)
CI_20us_dark2 = st.norm.interval(alpha=0.95, loc=79.44, scale=SEM_20us_dark2)
CI_200us_dark2 = st.norm.interval(alpha=0.95, loc=79.46, scale=SEM_200us_dark2)

#convert 95% CI into single distance for plotting input purposes
CI_true_dark_distance = (CI_true_dark[1]-CI_true_dark[0])/2

CI_20ns_light_distance = (CI_20ns_light[1]-CI_20ns_light[0])/2
CI_20us_light_distance = (CI_20us_light[1]-CI_20us_light[0])/2
CI_200us_light_distance = (CI_200us_light[1]-CI_200us_light[0])/2

CI_20ns_dark2_distance = (CI_20ns_dark2[1]-CI_20ns_dark2[0])/2
CI_20us_dark2_distance = (CI_20us_dark2[1]-CI_20us_dark2[0])/2
CI_200us_dark2_distance = (CI_200us_dark2[1]-CI_200us_dark2[0])/2

```

```python
plt.figure(figsize=(5,4), dpi=300)
plt.rcParams.update({'font.size': 16})
# plt.scatter(x=[20, 20000, 200000], y=[79.52, 79.52, 79.51], label="light", color="#b72818")
plt.errorbar(x=[20, 20000, 200000], y=[79.52, 79.52, 79.51], yerr=[CI_20ns_light_distance, CI_20us_light_distance, CI_200us_light_distance], capsize=3, fmt=".", color="#b72818", label='$F_{t}$')
# plt.scatter(x=[20, 20000, 200000], y=[79.44, 79.44, 79.46], label="dark2", color="grey")
plt.errorbar(x=[20, 20000, 200000], y=[79.44, 79.44, 79.46], yerr=[CI_20ns_dark2_distance, CI_20us_dark2_distance, CI_200us_dark2_distance], capsize=3, fmt=".", color="grey", label='$F_{dark2}$')
# plt.scatter(x=[1], y=[79.45], label="true dark", color="black")
plt.errorbar(x=[1], y=[79.45], yerr=[CI_true_dark_distance], capsize=3, fmt=".", color="black", label='$F_{laser\ off}$')
plt.plot([1, 20, 20000, 200000], [79.45, 79.52, 79.52, 79.51], color="#b72818", ls="--", lw=1)
plt.plot([1, 20, 20000, 200000], [79.45, 79.44, 79.44, 79.46], color="grey", ls="--", lw=1)
plt.xscale('log')
plt.ylabel("Crystallographic a/b Axes ($\AA$)")
plt.xlabel("Pump-Probe Time Delay (ns)")
plt.xlim(.2,1000000)
# plt.ylim(19.7,21)
plt.legend(loc='center', labelspacing=0.2, handletextpad=0.1, borderpad=0.2, borderaxespad=0.2)
plt.tight_layout()
```
